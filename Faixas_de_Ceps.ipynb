```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
```

```python
url_cco = 'https://raw.githubusercontent.com/rnisiide/ceps_brasil/main/CEPS%20Chapec%C3%B3%20CCO.csv'
url_cwb = 'https://raw.githubusercontent.com/rnisiide/ceps_brasil/main/CEPS%20CURITIBA.csv'
df_cco = pd.read_csv(url_cco)
df_cwb = pd.read_csv(url_cwb)

# Dataset is now stored in a Pandas Dataframe
```

```python
df_cco.head()
```

```python
## Para Chapecó
bairro_inicial = 'Centro'
cep_inicial = '89801-000'
bairro_final = ''
cep_final = ''
lista_faixas = []
nova_faixa = []
for index, row in df_cco.iterrows():
  if row["Bairro"] != bairro_inicial:
    cep_final = df_cco.loc[index-1,"CEP"]
    nova_faixa = [bairro_inicial, cep_inicial, cep_final]
    lista_faixas.append(nova_faixa)
    bairro_inicial = row["Bairro"]
    cep_inicial = row["CEP"]
    ##print(index, nova_faixa)
  elif (index == len(df_cwb)-1):
    nova_faixa = [row["Bairro"], cep_inicial, row["CEP"]]
    lista_faixas.append(nova_faixa)
    ##print(index, nova_faixa)
  else:
    continue

lista_faixas_df_cco = pd.DataFrame(lista_faixas, columns=['Bairro', 'CEP Inicial', 'CEP Final'])
##lista_faixas_df_cwb.to_csv('/content/drive/My Drive/Faixas Curitiba.csv', index=False)
lista_faixas_df_cco.head()
```
